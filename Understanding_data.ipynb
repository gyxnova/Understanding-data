# Multivarient

```python
tips = sns.load_dataset("tips")
```

```python
titanic = pd.read_csv("/content/tested.csv")
titanic
```

```python
iris = sns.load_dataset("iris")
```

```python
flights = sns.load_dataset("flights")
```

```python
# scatter plot
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex",style = "smoker",size="size",palette="magma")
plt.show()
```

```python
# Bar plot
sns.barplot(x="Pclass",y="Age",data=titanic,hue="Sex")
plt.show()
```

```python
sns.boxplot(x="Sex",y="Age",data=titanic,hue="Survived")
plt.show()
```

```python
sns.kdeplot(data=titanic[titanic["Survived"]==0]["Age"],fill=True)
sns.kdeplot(data=titanic[titanic["Survived"]==1]["Age"],fill=True)
plt.show()
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
plt.show()
```

```python
titanic.groupby("Embarked")["Survived"].mean()*100
```

```python
sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
plt.show()
```

```python
sns.pairplot(data=iris,hue="species")
plt.show()
```

```python
# Lineplot
sns.lineplot(x="year",y="passengers",data=flights)
plt.show()
```

```python
sns.heatmap(flights.pivot_table(index="month",columns="year",values="passengers"))
plt.show()
```

```python
sns.clustermap(flights.pivot_table(index="month",columns="year",values="passengers"))
plt.show()
```
